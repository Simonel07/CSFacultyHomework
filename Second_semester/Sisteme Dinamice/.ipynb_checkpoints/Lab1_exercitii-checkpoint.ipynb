# Exercitiul 8

a.

```python
t,s = var('t,s')
f(t,s) = 1 - (s*cos(4*t)*cos(t)/sqrt(1-s^2*cos(4*t)^2*sin(t)^2))
x(t) = f(t-pi/2, 0.5)
y(t) = f(t, 0.5)
parametric_plot((x(t),y(t)), (t,0,2*pi))
```

b.

```python
t,s = var('t,s')
f(t,s) = 1 - (s*cos(4*t)*cos(t)/sqrt(1-s^2*cos(4*t)^2*sin(t)^2))
x(t) = f(t-pi/2, 1/10)
y(t) = f(t, 1/10)
g = parametric_plot((x(t),y(t)), (t,0,2*pi))
for i in [2..10]:
    x(t) = f(t-pi/2, i/10)
    y(t) = f(t, i/10)
    g1 = parametric_plot((x(t),y(t)), (t,0,2*pi))
    g = g+g1
show(g)
```
