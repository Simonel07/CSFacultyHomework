# Laborator 2

Problema 1

a.

```python
reset()
x = var('x')
y = function('y')(x)
eqd = diff(y,x) == 2*x*(1+y^2)
sol = desolve(eqd,y)
sol_exp = solve(sol,y)
equ = sol_exp[0].rhs()
_C = var('_C')
sol1 = equ.substitute(_C==1)
show(eqd)
show(sol1)
plot(sol1,x,-3,7,ymin=-5,ymax=5, detect_poles = True)
```

b.

```python
reset()
x = var('x')
y = function('y')(x)
eqd = (x^2-1)*diff(y,x)+2*x*y^2 == 0
eqd
```

```python
sol = desolve(eqd,y)
sol_exp = solve(sol,y)
equ = sol_exp[0].rhs()
_C = var('_C')
sol1 = equ.substitute(_C==1)
show(sol1)
plot(sol1,x,5,100)
```

```python
reset()
x = var('x')
y = function('y')(x)
eqd = diff(y,x) == -((x+y)/y)
eqd
```
