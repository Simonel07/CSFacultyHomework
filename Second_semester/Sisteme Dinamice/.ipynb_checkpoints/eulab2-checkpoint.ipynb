```python
reset()
_C = var("_C")
x = var('x')
k = var('k')
y = function('y')(x)
eqd = diff(y, x) + (k/x)*y == x^3
sol(x, k, _C) = desolve(eqd, [y, x])
show(sol(x, k, _C))
show(sol(x, 1, _C))
lst = [sol(x, 1, i) for i in [10..20]]
plot(lst, x, -4, 4, detect_poles = True, ymax=40, ymin=-40)
```
