```python
t = var('t')
k = var('k')
x0 = var('x0')
x(t) = function('x')(t)
eqd = diff(x,t) == -k*x
f(t, x0, k) = desolve(eqd, [x, t], ics = [0, x0])
show(f(t, x0, k))
```

```python
g = plot(f(t, 10/1, 0.1), t, 0, 100)
for i in range(1, 10):
    g += plot(f(t,10/i, 0.1), t, 0, 100)
show(g)
```

```python
t = var('t')
k = var('k')
Tm = var('Tm')
T0 = var('T0')
T(t) = function('T')(t)
eqd = diff(T, t) == -(k*(T(t) - Tm))
f(t, k, Tm, T0) = desolve(eqd, [T, t], ics=[0, T0])
show(f(t, k, T0, Tm))
```

```python
p = plot(f(t, 1, 21, 36), t, 0, 5)
cc = ['red', 'blue', 'green', 'yellow', 'brown']
for i in [2..10]:
    p += plot(f(t, i, 21, 36), t, 0, 5, color=cc[i%5])
show(p)
```
