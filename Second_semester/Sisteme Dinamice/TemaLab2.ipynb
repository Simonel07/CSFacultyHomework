Problema 4

```python
reset()
x = var('x')
a = var('a')
b = var('b')
y = function('y')(x)
eqd = diff(y,x) == y + a*x + b
sol(x) = desolve(eqd, dvar=[y,x], ics = [0,1])
syst = [sol(1) == -5 + 4*e, sol(2) == -7 + 4*e^2]
solve(syst, a, b)
```

```python
solg = sol.substitute(a == 2, b == 1)
plot(solg, x, -10, 10, ymin = -20, ymax = 20)
```

```python
reset()
x = var('x')
a = var('a')
y = var('y')
eqd = diff(y,x,2)-diff(y,x)-2*y==0
sol = desolve(eqd, y, ics=[0,a,2])
```

```python
show(sol)
assume(a>-2)
limit(sol, x=Infinity)
forget()
assume(a<-2)
limit(sol, x=Infinity)
forget()
sol1 = sol.substitute(a==-2)
limit(sol1, x=infinity)
```
